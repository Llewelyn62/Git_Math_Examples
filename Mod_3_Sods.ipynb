```python
#Initial values
nx = 81
dx = .25
dt = 0.0002
nt = 50 # This function appears to run out to around 1800, where v and pressure are stable and density is almost even. 3.6seconds. 
#Set up an initial array of values
U = np.ones((3, nx)) # An array of three variables (density, velocity & energy) x distance.
#Initial values
rho_L, rho_R = 1, 0.125
vel_L, vel_R = 0, 0
p_L, p_R = 100000,10000
gamma = 1.4
e_L = p_L/((gamma-1)*rho_L)
e_R = p_R/((gamma-1)*rho_R)
e_T_L = e_L+(vel_L**2/2.)
e_T_R = e_R+(vel_R**2/2.)
#Calculate the initial values in the tube for the first row
U[0,0:41] = rho_L
U[1,0:41] = rho_L*vel_L #velocity
U[2,0:41] = rho_L*e_T_L #energy
U[0,41:] = rho_R #density
U[1,41:] = rho_R*vel_R #velocity
U[2,41:] = rho_R*e_T_R #energy
np.shape(U)
U_move = U.copy()
```

```python
#Set up functions -- these need to take in some variables and output an array (3) reflecting the vectors f and u.
def computeF(u_value):
    u1 = u_value[0] #density
    u2 = u_value[1] #velocity
    u3 = u_value[2] #energy
    flux = np.empty_like(u_value)
    flux[0] = u2
    flux[1] = (u2**2/u1)+(gamma-1)*(u3-u2**2/(2*u1))
    flux[2] = (u3+(gamma-1)*(u3-u2**2/(2*u1)))*u2/u1
    
    return flux
```

```python
def Richtmyer(U,nt,dx,dt):
    UN_plus = np.empty_like(U) #Create two arrays for results of half-points
    UN_minus = np.empty_like(U)
    UN = np.empty_like(U)
    
    for i in range(nt):
        #Richtmyer
        #Step 1: Create the half-step arrays.
        UN_plus[:,:-1] = (U[:,:-1] + U[:,1:])/2 - (dt/(2*dx))*(computeF(U[:,1:])-computeF(U[:,:-1]))
        UN_plus[:,-1] = UN_plus[:,-2]
        UN_minus[:,1:] = UN_plus[:,:-1]
        UN_minus[:,0] = UN_minus[:,1]
        #Step 2: Calculate full-step value .
        #UN+ & UN- move array inward at each end, so need 1:-1.
        #The next line needs to start at U[:,1] as there cannot be a reasonable u minus half outside of the domain. 
        #Similarly, it must end before an artifical u plus 1 is added at the other end. 
        #UN_plus must start at the 1 index, as that is the value for U[:,1] and U[:,2]
        UN[:,1:-1] = U[:,1:-1] - (dt/dx)*(computeF(UN_plus[:,1:-1])-computeF(UN_minus[:,1:-1]))
        #Tidy the boundariy conditions by setting them equal to the prior value, hence avoiding Dirichlet boundary issues. 
        UN[:,0] = UN[:,1]
        UN[:,-1] = UN[:,-2]
        #Copy the temp updated-U over to result U, no index shift. 
        U[:,:] = UN.copy()

        return U
```

```python
def Convert_back(U):
    rho = U[0,:]
    velocity = U[1,:]/U[0,:]
    pressure = (gamma - 1)*(U[2,:] - U[1,:]**2/(2.*U[0,:]))
    return rho, pressure, velocity

Richtmyer(U,nt,dx,dt)
round(rho[51],2), round(velocity[51],2), round(pressure[51],2)
```
