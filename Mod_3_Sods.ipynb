# Sod's Test

```python
#Set up modules
import numpy as np
```

```python
#Initial values
nx = 81
dx = .25
dt = 0.0002

nt = 50
#Set up an initial array of values
U = np.ones((3, nx)) # An array of three variables (density, velocity & energy) x distance.
#Initial values
rho_L, rho_R = 1., 0.125
vel_L, vel_R = 0, 0
p_L, p_R = 100000.,10000.
gamma = 1.4
e_L = p_L/((gamma-1)*rho_L)
e_R = p_R/((gamma-1)*rho_R)
e_T_L = e_L+(vel_L**2/2.)
e_T_R = e_R+(vel_R**2/2.)
#Calculate the initial values in the tube for the first row
U[0,0:(nx-1)/2] = rho_L
U[1,0:(nx-1)/2] = rho_L*vel_L #velocity
U[2,0:(nx-1)/2] = rho_L*e_T_L #energy
U[0,(nx-1)/2 :] = rho_R #density
U[1,(nx-1)/2 :] = rho_R*vel_R #velocity
U[2,(nx-1)/2 :] = rho_L*e_T_R #energy
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-10, 10,nx)
plt.ylim(0,1.1*max(U[0,:]))
plt.plot(x,U[0,:], color='#003366', ls='-', lw=3)
plt.show()
```

```python
#Set up functions -- these need to take in some variables and output an array (3) reflecting the vectors f and u.
def computeF(u_value):
    u1 = u_value[0] #density
    u2 = u_value[1] #velocity
    u3 = u_value[2] #energy
    flux = np.empty_like(u_value)
    flux[0] = u2
    flux[1] = (u2**2/u1)+(gamma-1)*(u3-u2**2/(2*u1))
    flux[2] = (u3+(gamma-1)*(u3-u2**2/(2*u1)))*u2/u1
    
    return flux
```

```python
UN_plus = np.empty_like(U) #Create two arrays for results of half-points
UN_minus = np.empty_like(U)
UN = np.empty_like(U)
for i in range(nt):
    #Richtmyer
    #Step 1: Create the half-step arrays.
    UN_plus[:,:-1] = 1./2.*(U[:,:-1] + U[:,1:]) - (dt/(2.*dx))*(computeF(U[:,1:])-computeF(U[:,:-1]))
    UN_plus[:,-1] = UN_plus[:,-2]
    UN_minus[:,1:] = UN_plus[:,:-1]
    UN_minus[:,0] = UN_minus[:,1]
    #Step 2: Calculate full-step value .
    #UN+ & UN- move array inward at each end, so need 1:-1.
    #The next line needs to start at U[:,1] as there cannot be a reasonable u minus half outside of the domain. 
    #Similarly, it must end before an artifical u plus 1 is added at the other end. 
    #UN_plus must start at the 1 index, as that is the value for U[:,1] and U[:,2]
    UN[:,1:-1] = U[:,1:-1] - (dt/dx)*(computeF(UN_plus[:,1:-1])-computeF(UN_minus[:,1:-1]))
    #Tidy the boundariy conditions by setting them equal to the prior value, hence avoiding Dirichlet boundary issues. 
    UN[:,0] = UN[:,1]
    UN[:,-1] = UN[:,-2]
    #Copy the temp updated-U over to result U, no index shift. 
    U[:,:] = UN.copy()
```

```python
rho = U[0,:]
velocity = U[1,:]/U[0,:]
pressure = (gamma - 1)*(U[2,:] - U[1,:]**2/(2.*U[0,:]))
rho[50],velocity[50],pressure[50], max(pressure)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 14, 6
x = np.linspace(-10, 10,nx)

fig, (ax0, ax1,ax2) = plt.subplots(ncols=3)
fig.suptitle('Shock-tube results $t=.01$', fontsize=16)
ax0.grid(True)
ax0.plot(x, rho[:],color='#003366', ls='-', lw=3)
ax0.set_title('Density')
ax0.set_ylim(min(rho)*.9,max(rho)*1.1)
ax1.grid(True)
ax1.plot(x, velocity[:],color='#003366', ls='-', lw=3)
ax1.set_title('Velocity')
ax1.set_ylim(min(velocity)-.1,max(velocity)*1.1+.2)
ax2.grid(True)
ax2.plot(x, pressure[:],color='#003366', ls='-', lw=3)
ax2.set_title('Pressure')
ax2.set_ylim(min(pressure)-1,max(pressure)*1.1+1)

# Tweak spacing between subplots to prevent labels from overlapping
plt.subplots_adjust(hspace=0.5)

plt.show()
fig.savefig('velocity.png',dpi=200, format='png')
```
